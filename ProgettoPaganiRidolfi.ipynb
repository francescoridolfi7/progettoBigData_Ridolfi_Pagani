## <font color="green"> Informazioni sul dataset: </font>

Questo dataset contiene informazioni sugli incidenti stradali in India dal 2014 al 2016, come il nome dello stato e della regione dove si sono registrati, le cause di incidenti (come assunzione di alchool, eccesso di velocità , sorpassi, invasione di corsia , colpi di sonno ecc.)

## <font color="green"> Obiettivo di ricerca: </font>

L'obiettivo dell'analisi di questo dataset è quello di comprendere al meglio le cause di incidenti stradali e di conseguenza:
- Installare misure preventive (come aggiunta di segnaletica stradale, incrementare la presenza di vigili o pattuglie stradali sul territorio, uso di nuove tecnologie per rilevare velocità ecc.)
- Identificare se il rispetto delle norme di sicurezza è peggiorato o migliorato nel tempo 


## <font color="green"> Fonte dati: </font>

Il dataset da noi utilizzato si trova al seguente link: 
https://www.kaggle.com/datasets/thedevastator/state-ut-wise-road-accidents-due-to-driver-viola


## <font color="yellow"> Analisi dati: </font>

```python
#importazione delle librerie
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Caricamento dataset: 

```python
#caricamento e lettura del dataframe
df = pd.read_csv('driverresponse.csv')
df
```

### <font color="yellow"> Classifica stati con più incidenti: </font>
Raggruppiamo gli stati in base al numero di incidenti, questo ci aiuterà a capire le aree in cui bisogna intervenire tempestivamente e introdurre miglioramenti per la sicurezza

#### Nel 2014:


```python
#creazione lista contenente tutte le tipologie di incidenti
incidenti2014 = ['alcintake2014','overspeed2014','overtaking2014','lanejumping2014','wrongside2014','signalavoid2014','asleep2014','othercause2014']

#creazione di colonna contentente il totale degli incidenti nel 2014
df['Totale_2014'] = df[incidenti2014].sum(axis=1)
colonne = list(df.columns)

#visualizzazione delle colonne regione stato e totale incidenti ordinati in maniera decrescente
df.loc[:, ["stateut", "region", "Totale_2014"]].sort_values('Totale_2014', ascending=False)
```

#### Nel 2016: 

```python
#creazione colonna per totale incidenti nel 2016, attraverso la somma di tutte le tipologie di incidenti
df['Totale_2016'] = df['alcintake2016'] + df['asleep2016'] + df['lanejumping2016'] + df['overspeed2016'] + df['overtaking2016'] + df['signalavoid2016'] + df['wrongside2016'] + df['othercause2016']

#visualizzazione delle colonne regione stato e totale incidenti ordinati in maniera decrescente
df.loc[:, ["stateut", "region", "Totale_2016"]].sort_values('Totale_2016', ascending=False)
```

```python
#stampa del grafico a barre per visualizzare il confronto tra incidenti totali nei due diversi anni (per ogni stato)
df.plot.bar(x='stateut',y=['Totale_2014','Totale_2016'])
```

### <font color="yellow"> Incidenti per tipologia </font>
Abbiamo deciso di mostrare due grafici a torta  uno per ogni anno, per visualizzare il confronto tra le varie tipologie di incidenti per vedere quelle più frequenti e come esse siano cambiate in 2 anni

#### Nel 2014:

```python
#creazione variabili contenenti la somma degli incidenti per ogni tipologia
sumAlc2014 = df['alcintake2014'].sum()
sumSpd2014 = df['overspeed2014'].sum()
sumTkg2014 = df['overtaking2014'].sum()
sumJmp2014 = df['lanejumping2014'].sum()
sumWrgSide2014 = df['wrongside2014'].sum()
sumSgl2014 = df['signalavoid2014'].sum()
sumSleep2014 = df['asleep2014'].sum()
sumOther2014 = df['othercause2014'].sum()

#gestione stile del grafico
plt.style.use("ggplot")
slices = [sumAlc2014, sumJmp2014, sumSpd2014, sumTkg2014, sumWrgSide2014, sumSgl2014, sumSleep2014, sumOther2014]
#creazione etichette e colori per ogni tipologia di incidente
labels = ['Alchool', 'Salto di corsia','Eccesso velocità', 'Sorpasso', 'Contro Mano', 'Segnaletica Non Rispettata', 'Sonno Conducente', 'Altro']
colors = ['orange', 'brown', 'red', 'green', 'blue', 'yellow', 'purple', 'pink']
plt.pie(slices, autopct='%1.1f%%',pctdistance = 1.2, colors=colors)
plt.title("Tipologie Incidenti 2014")
#creazione legenda per il grafico
plt.legend(labels, loc='best', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)
plt.show()
```

#### Nel 2016:

```python
#creazione variabili contenenti la somma degli incidenti per ogni tipologia
sumAlc2016 = df['alcintake2016'].sum()
sumSpd2016 = df['overspeed2016'].sum()
sumTkg2016 = df['overtaking2016'].sum()
sumJmp2016 = df['lanejumping2016'].sum()
sumWrgSide2016 = df['wrongside2016'].sum()
sumSgl2016 = df['signalavoid2016'].sum()
sumSleep2016 = df['asleep2016'].sum()
sumOther2016 = df['othercause2016'].sum()

#gestione stile del grafico
plt.style.use("ggplot")
slices = [sumAlc2016, sumJmp2016, sumSpd2016, sumTkg2016, sumWrgSide2016, sumSgl2016, sumSleep2016, sumOther2016]
#creazione etichette e colori per ogni tipologia di incidente
labels = ['Alchool', 'Salto di corsia','Eccesso velocità', 'Sorpasso', 'Contro Mano', 'Segnaletica Non Rispettata', 'Sonno Conducente', 'Altro']
colors = ['orange', 'brown', 'red', 'green', 'blue', 'yellow', 'purple', 'pink']
plt.pie(slices, autopct='%1.1f%%',pctdistance = 1.2, colors=colors)
plt.title("Tipologie Incidenti 2016")
#creazione legenda per il grafico
plt.legend(labels, loc='best', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)
plt.show()
```

### <font color="yellow"> Incidenti per ogni regione divisi per tipologie: </font>
Abbiamo deciso di utilizzare un grafico a barre per determinare i tipi di violazioni più comunemente osservati regione per regione.

```python
#creazione dimensione grafico
plt.rcParams['figure.figsize'] = (6, 6)
#creazione variabile "fig" dove raggruppiamo le tipologie divise per regione
fig= df.groupby('region')[['alcintake2016','overspeed2016','overtaking2016', 'lanejumping2016','wrongside2016', 'signalavoid2016', 'asleep2016', 'othercause2016']].agg('sum')
labels = ['Alchool','Eccesso velocità', 'Sorpasso', 'Salto di corsia','Contro Mano', 'Segnaletica Non Rispettata', 'Sonno Conducente', 'Altro']
fig.plot(kind='bar',title='Totale incidenti per tipologia 2016')
plt.legend(labels, loc='best', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)
```

## <font color="pink"> Regressione lineare: </font>

Abbiamo deciso di effettuare una regressione lineare su due casi di studio per gli incidenti avvenuti nel 2016 in India.


```python
#importazione librerie necessarie per effettuazione regressione lineare
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#### Regressione linare in base ad incidenti per guida in stato di ebrezza:

```python
# Vogliamo effettuare una regressione lineare sulla correlazione tra gli incidenti con bassa percentuale, ed il totale di incidenti per l'anno 2016
X = df['alcintake2016'].to_numpy().reshape(-1,1)
y = df['Totale_2016'].to_numpy()

#inserimento di label per i valori su asse x ed y
plt.xlabel("Incidenti per stato di ebrezza")
plt.ylabel("Totale incidenti")
plt.scatter(X, y)
```

```python
# Separazione dei dataset di train e di test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
plt.scatter(X_train, y_train, label='train set')
plt.scatter(X_test, y_test, label='test set')

plt.xlabel("Incidenti per stato di ebrezza")
plt.ylabel("Totale incidenti")

plt.legend()
plt.show()
```

```python
# Creazione oggetto per la regressione lineare
reg = LinearRegression()

# Creazione train model con i due parametri train
reg.fit(X_train, y_train)

# Creare previsioni utilizzando il set di test
y_pred = reg.predict(X_test)

plt.scatter(X_test, y_test, label='test set')
plt.plot(X_test, y_pred, color='red', label='predicted values')

plt.xlabel("Incidenti per stato di ebrezza")
plt.ylabel("Totale incidenti")

plt.legend()
plt.show()
```

#### Regressione linare in base ad incidenti per eccesso di velocità:

```python
# Vogliamo effettuare una regressione lineare sulla correlazione tra gli incidenti con più percentuale, ed il totale di incidenti per l'anno 2016
X = df['overspeed2016'].to_numpy().reshape(-1,1)
y = df['Totale_2016'].to_numpy()

plt.xlabel("Incidenti per eccesso di velocità")
plt.ylabel("Totale incidenti")

plt.scatter(X, y)
```

```python
# Divisione dei dataset di train e di test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
plt.scatter(X_train, y_train, label='train set')
plt.scatter(X_test, y_test, label='test set')

plt.xlabel("Incidenti per eccesso di velocità")
plt.ylabel("Totale incidenti")

plt.legend()
plt.show()
```

```python
# Creazione regressione lineare
reg = LinearRegression()

# Creazione train model con i due parametri train
reg.fit(X_train, y_train)

# Creazione di previsioni con i test
y_pred = reg.predict(X_test)

plt.scatter(X_test, y_test, label='test set')
plt.plot(X_test, y_pred, color='red', label='predicted values')

plt.xlabel("Incidenti per eccesso di velocità")
plt.ylabel("Totale incidenti")

plt.legend()
plt.show()
```

### <font color="green"> Conclusioni sulla regressione lineare eseguita: </font>
Grazie alla regressione lineare effettuata su due tipologie diverse di incidenti risulta subito chiaro che gli incidenti per eccesso di velocità nel 2016 hanno avuto un impatto molto significativo sul totale di incidenti rispetto ad esempio agli incidenti per guida in stato di ebrezza. <br>
I grafici evidenziano il fatto che bisogna intervenire tempestivamente con l'installazione di autovelox , tutor o aumentare il numero di pattuglie e/o posti di blocco, per diminuire quanto possibile il numero di incidenti per eccesso di velocità.
